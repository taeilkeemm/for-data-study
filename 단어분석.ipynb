```python
from konlpy.tag import Okt
from collections import Counter
import csv
```

```python
title = ''
  
for i in range(0, 2500):
  title = title + ' ' + df.loc[i][0]
```

```python
title
```

```python
okt = Okt()
noun = okt.nouns(title)
for i, v in enumerate(noun):
  if len(v) < 2:
    noun.pop(1)
```

```python
count = Counter(noun)
```

```python
noun_list = count.most_common(400)
for v in noun_list:
  print(v)
```

```python
import codecs
```

```python
positive = []
negative = []
posneg = []
pos = codecs.open("pos.txt")
```

```python
while True:
  line = pos.readline()
  line = line.replace('\n', '')
  positive.append(line)
  posneg.append(line)
  if not line:
    break
pos.close()
```

```python
neg = codecs.open("neg.txt")
```

```python
while True:
  line = neg.readline()
  line = line.replace('\n', '')
  negative.append(line)
  posneg.append(line)
  if not line:
    break
neg.close()
```

```python
from tqdm import tqdm
import re
```

```python
labels = []
title_data = list(df['뉴스제목'])
```

```python
positive.pop()
negative.pop()
```

```python
negative
```

```python
for title in tqdm(title_data):
  clean_title = re.sub('[-=+,#/\?:^$.@*\"※~&%ㆍ!』\\‘|\(\)\[\]\<\>`\'…\"\“》]', '', title)
  negative_flag = False
  label = 0
  for i in range(len(negative)):
    if negative[i] in clean_title:
      label =  -1
      negative_flag = True
      break
  if negative_flag == False:
    for i in range(len(positive)):
      if positive[i] in clean_title:
        label = 1
        break
  labels.append(label)

df['label'] = labels
```
